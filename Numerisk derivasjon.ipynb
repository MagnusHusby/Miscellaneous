# Hva er derivasjon?

Derivasjon er stigningen i en funksjon. 
En måte og se for seg den deriverte i et punkt er ved å tegne en sekant på grafen til en funksjon, og la de to skjæringspunktene bevege seg mot hverandre til de er uendelig nærme hverandre på dette punktet. Da vil du ha en tangent, en linje som kun skjærer grafen kun ett sted, og stigningstallet dens er da den deriverte i punktet.

I matematikk definerer vi den deriverte slik:
$$ f'(x) = \lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x} = \lim_{\Delta x \to 0}\frac{\Delta y}{\Delta x}$$


### Numerisk derivasjon

I motsetning til å finne funksjonen til den deriverte av en funksjon, er numerisk derivasjon heller bare det å finne hvert enkelt stigningstall til tangentene på mange punkter på en funksjon. 

Det er dette en PC gjør når den regner ut den deriverte for deg.

Her er et eksempel på numerisk derivasjon av $f(x) = x^2$:

```python
from pylab import *

def eksfunk(x):
    y = x**2
    return y

y_verdi = []
x_verdi = []

for i in range(11):
    x_verdi.append(i)          # Fyller listen med x-verdier
    y_verdi.append(eksfunk(i)) # Finner funksjonsverdien for alle x-ene

#print(y_verdi)
#print(x_verdi)
    
plot(x_verdi, y_verdi)
title("Min eksempelfunksjon")
show()

y_deriv = [0]

for i in range(10):  # Den må finne den deriverte i alle punktene for x. Her er det bare 10 for den plusser på en i funksjonen
    deriv = (y_verdi[i+1]-y_verdi[i])/(x_verdi[i + 1]- x_verdi[i])  # Her regner den ut delta y delt på delta x flere ganger
    y_deriv.append(deriv)

#print(y_deriv)
plot(x_verdi, y_deriv)
title("Den deriverte av min eksempelfunksjon")
show()
```

### Diskré derivasjon

Eksempel fra klasserommet:

```python
from pylab import *

t = [5, 10, 15, 30, 50, 60]           # Tidspunkter i sekunder
s = [115, 190, 390, 600, 1000, 1200]  # Ernest sin avstand fra start ved tilsvarende tidspunkt
v = []                                # En liste med tilsvarende hastighet
v.append(0)

for i in range(len(s)-1):
    sder = (s[i+1] - s[i])/(t[i+1]-t[i])
    v.append(sder)

plot(t, s, color = "blue")
xlabel("Tid i sekunder")
ylabel("Avstand i cm")
title("Ernest som går gjennom klasserommet, imens Aksel desperat prøver å måle.")
grid()
show()

plot(t, v, color = "red")
xlabel("Tid i sekunder")
ylabel("Hastigheten i cm/s")
title("Hastigheten til Ernest.")
grid()
show()
```

Eksempel med måling av heisens bevegelse:

```python
from pylab import *

data = loadtxt('height.txt', delimiter = ',', skiprows = 1)

tid = data [:,0]
hoyde = data [:,2]


plot(tid,hoyde, color = "red")
title("Heisens høyde fra bunnen")
xlabel("Tid i sekunder")
ylabel("Høyde i meter")
grid()
show()

v = [0]


for i in range(len(tid)-1):
    fart = (hoyde[i+1]-hoyde[i])/(tid[i + 1]- tid [i])
    v.append(fart)
      
plot(tid,v, color = "green")
title("Heisens hastighet oppover")
xlabel("Tid i sekunder")
ylabel("Hastighet i m/s")
grid()
show()

a= [0]

for i in range(len(tid)-1):
    akselerasjon = (v[i+1]-v[i])/(tid[i + 1]- tid [i])
    a.append(akselerasjon)
    
plot(tid, a, color = "blue")
title("Heisens akselerasjon")
xlabel("Tid i sekunder")
ylabel("Akselerasjon i m/s^2")
grid()
show()
```

### Derivasjon med SymPy

Og hvis du ikke gidder å derivere numerisk, så kan man finne uttrykket for den derivert også ved hjelp av SymPy.
Her må du først importere SymPy-biblioteket og kalle det noe. Jeg kalte det for "sym". Så må jeg definere at "x" er ukjent ved "sym.Symbol("x")"
Da kan jeg bruke funksjonen "sym.diff()"
- Den trenger kun en funksjon av noe slag som input og vil så gi deg den funksjonens deriverte.

Her er et eksempel på veldig simpel derivasjon av et uttrykk:

```python
import sympy as sym
from pylab import *

x = sym.Symbol("x")

deriv_funk = sym.diff(x**2 + 7*x + 4, x)

print(deriv_funk)
```
